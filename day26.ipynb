```python
# polynomial regression
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# data
df = pd.DataFrame({
    "Experience" :[1,2,3,4,5]
})

x = df
y = [10,20,50,90,150]

# poly feature
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
print(x_poly)


# model
model = LinearRegression()
model.fit(x_poly,y)

# prediction
input_data = int(input('Enter the experience:'))
new_data = pd.DataFrame({
    'Experience':[input_data]
})

# convert new data to poly
u_new_data = poly.fit_transform(new_data)
predicted_data = model.predict(u_new_data)
print(predicted_data)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Data
df = pd.DataFrame({
    "weight": [120, 140, 150, 170],
    "size": [6, 7, 7, 8]
})

# Features
x = df

# Target
y = ["Apple", "Apple", "Orange", "Orange"]

# Model
model = DecisionTreeClassifier()
model.fit(x, y)

# Prediction Input
input_weight = int(input("Enter the weight: "))
input_size = int(input("Enter the size: "))

new_data = pd.DataFrame({
    "weight": [input_weight],
    "size": [input_size]
})

# Prediction
predicted_data = model.predict(new_data)

print("Predicted Fruit:", predicted_data[0])
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt


# Dataset
df = pd.DataFrame({
    "Marks": [20, 35, 40, 50, 60, 75, 90]
})

# Features
x = df

# Target
y = ["Fail", "Fail", "Pass", "Pass", "Pass", "Pass", "Pass"]

# Model
model = DecisionTreeClassifier()
model.fit(x, y)

# User Input
marks = int(input("Enter your marks: "))

new_data = pd.DataFrame({
    "Marks": [marks]
})

# Prediction
result = model.predict(new_data)

print("Result:", result[0])
# Decision Tree Graph
plt.figure(figsize=(8,5))

plot_tree(
    model,
    feature_names=["Marks"],
    class_names=["Fail", "Pass"],
    filled=True,
    rounded=True
)

plt.show()
```
