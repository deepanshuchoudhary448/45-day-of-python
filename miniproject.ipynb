```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler , LabelEncoder
url = "https://raw.githubusercontent.com/yashmourya236-code/Data_Science_-45-Days-/main/moviesTMBD.csv"
df = pd.read_csv(url)
print(df)
print('\n')

#to clean the data
df.drop(["id" , "title" , "adult" , "original_language" , "overview" , "release_date" , "Unnamed: 0"], axis = 1 ,inplace = True)
print(df)



#feature scaling
scaler = StandardScaler()
df[["vote_average","vote_count"]] = scaler.fit_transform(df[["vote_average","vote_count"]])
print(df)


#split data
x = df[["vote_average" ,"vote_count"]]
y = df[["popularity"]]


#train_test_split
x_train ,x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)

print(x_train)
print('\n')

print(x_test)
print('\n')

print(y_train)
print('\n')

print(y_test)
print('\n')

#model training
model = LinearRegression()
model.fit(x_train , y_train)

#pridiction
n_vote = int(input("Enter No. Of Votes : "))
avg_vote = float(input("Enter No. Of Avg Votes : "))

New_data =pd.DataFrame(
     {
    "vote_average" : [avg_vote],
    "vote_count" : [n_vote]
}
)
# New_df = pd.DataFrame(New_data)
New_data[["vote_average","vote_count"]] = scaler.fit_transform(New_data[["vote_average","vote_count"]])

predicted_pop = model.predict(New_data)
print("The Popularity Will Be -> " ,predicted_pop[0])






isme 6th line me apna url dalna , aur same file download kari thi na movie wali
```
