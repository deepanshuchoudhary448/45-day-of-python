```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
df = pd.DataFrame({'color':['red','blue','green','red','blue']})
# sparse_output=False returns a dense numpy array ; set_output returns a pamdas dataframe
encoder = OneHotEncoder(sparse_output=False).set_output(transform="pandas")
df_encoded = encoder.fit_transform(df[['color']])
print(df_encoded)
```

```python
# label encoding
#one hot encoding
import pandas as pd
data = {
'colors':['red','blue','green','red','blue']
}

df = pd.DataFrame(data)
print("original data")
print(df)

# one hot encoding
encoded_df = pd.get_dummies(df,columns=['colors'])

print("after one hot encoding")
print(encoded_df)
```
