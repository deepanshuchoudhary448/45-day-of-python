```python
import matplotlib.pyplot as plt

# Data for graph 1
dairy_year = [2010, 2015, 2020, 2025]
dairy = [100, 520, 630, 400]

# Data for graph 2
farming_year = [1990, 2000, 2005, 2010]
farming = [300, 200, 250, 100]

# Create 2x2 subplot grid
fig, aux = plt.subplots(2, 2, figsize=(8, 6))

# Graph 1: Dairy production
aux[0][0].plot(dairy_year, dairy, marker='o')
aux[0][0].set_xlabel("Year")
aux[0][0].set_ylabel("Dairy")
aux[0][0].set_title("Dairy Production Graph")

# Graph 2: Farming production
aux[1][1].plot(farming_year, farming, marker='o', color='green')
aux[1][1].set_xlabel("Year")
aux[1][1].set_ylabel("Farming")
aux[1][1].set_title("Farming Production Graph")

# Optional: hide unused subplots
aux[0][1].axis('off')
aux[1][0].axis('off')

plt.tight_layout()

# Save first
plt.savefig("subplt.jpg")

# Then display
plt.show()

print(plt.gcf().canvas.get_supported_filetypes())
```

```python
import matplotlib.pyplot as plt

# Data
dairy_year = [2010, 2015, 2020, 2025]
dairy = [100, 520, 630, 400]

farming_year = [1990, 2000, 2005, 2010]
farming = [300, 200, 250, 100]

fig, aux = plt.subplots(2, 2, figsize=(10, 8))

# Graph 1 (top-left)
aux[0][0].plot(dairy_year, dairy)
aux[0][0].set_title("Dairy Production")

# Graph 2 (top-right)
aux[0][1].plot(farming_year, farming)
aux[0][1].set_title("Farming Production")

# Graph 3 (bottom-left)
aux[1][0].bar(dairy_year, dairy, color="orange")
aux[1][0].set_title("Dairy Bar Chart")

# Graph 4 (bottom-right)
aux[1][1].bar(farming_year, farming, color="green")
aux[1][1].set_title("Farming Bar Chart")

plt.tight_layout()
plt.savefig("subplt.jpg")
plt.show()

plt.gcf().canvas.get_supported_filetypes()
plt.savefig("subplt.jpg")
```
