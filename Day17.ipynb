```python
# pip install matpltolib
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025] # x chord
y = [100,200,250,300] # y chord
plt.plot(x,y) # Line grap
plt.xlabel("years")
plt.ylabel("sales")
plt.title("sales report")
plt.show() # grap show


```

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]

# graph size
plt.figure(figsize = (2,2))
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]

plt.plot(x,y,color= "green",marker ="* ",linestyle="dashed",linewidth = 4, markersize=14,)
plt.show()
```

```python

import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y1 = [100,200,250,300]
y2 = [150,185,195,300]
plt. plot(x,y1,label= "jeans")
plt.plot(x,y2,label="shirt")
plt.legend()
plt.show()
```
